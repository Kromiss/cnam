# Affichages

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()
plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.xlabel('Angle (rad)')
plt.ylabel('Valeur')
plt.legend()
plt.show()
```

```python
ir = np.array([1.0, 0.75, 0.5, 0.0, -0.25, 0.0, 0.125])
plt.stem(ir)
plt.show()
```

```python
n = 1000
b = 100
plt.figure()
plt.hist(np.random.uniform(size=n),bins=b)
plt.title('uniform')

plt.figure()
plt.hist(np.random.normal(size=n),bins=b)
plt.title('normal')

plt.figure()
plt.hist(np.random.laplace(size=n),bins=b)
plt.title('laplace')
plt.show()

```

```python
x = np.linspace(0,4*np.pi, 50)
sn = np.sin(x)
cs = np.cos(x)

fig, axs = plt.subplots(2, sharex=True)
axs[0].plot(x,sn)
axs[0].set_title('sin')
axs[1].plot(x,cs,color='red', marker='o')
axs[1].set_title('cos')
plt.show()
```

```python
n = 25
m = np.arange(n).reshape((5, 5))
print(m)
plt.imshow(m)
plt.colorbar()
plt.show()
n = 40
x = np.cos(np.linspace(0, 8*np.pi, n))
y = np.cos(np.linspace(0, 4*np.pi, n))
z = np.matrix(x).T * np.matrix(y)
plt.figure()
plt.imshow(z, aspect='auto', interpolation='bicubic')
plt.show()
```
